```python
%%manim -pqh spiral

from manim import *

class spiral(MovingCameraScene):
    def construct(self):
        s=Square(side_length=1)
        a=Arc(radius=1,start_angle=-PI/2,angle=PI/2,arc_center=s.get_corner(UL))
        s2 = Square(side_length=1).next_to(s,UP,buff=0)
        a2=Arc(radius=1,start_angle=-PI/2+PI/2,angle=PI/2,arc_center=s2.get_corner(DL))
        s3= Square(side_length=2).next_to(VGroup(s,s2),LEFT,buff=0)
        a3=Arc(radius=s3.side_length,start_angle=-PI/2+PI/2+PI/2,angle=PI/2,arc_center=s3.get_corner(DR))
        s4= Square(side_length=3).next_to(s3,DOWN,buff=0).align_to(s3,LEFT)
        a4=Arc(radius=s4.side_length,start_angle=-PI/2+PI/2+PI/2+PI/2,angle=PI/2,arc_center=s4.get_corner(UR))
        s5= Square(side_length=5).next_to(s4,RIGHT,buff=0).align_to(s4,DOWN)
        a5=Arc(radius=s5.side_length,start_angle=-PI/2+PI/2+PI/2+PI/2+PI/2,angle=PI/2,arc_center=s5.get_corner(UL))
        s6= Square(side_length=8).next_to(s5,UP,buff=0).align_to(s5,RIGHT)
        a6=Arc(radius=s6.side_length,start_angle=-PI/2+PI/2+PI/2+PI/2+PI/2+PI/2,angle=PI/2,arc_center=s6.get_corner(DL))
        texts=[1,1,2,3,5,8]
        texts2=VGroup()
        for k in [a,a2,a3,a4,a5,a6]:
            k.set_color(BLUE)
        for t,sq in zip(texts,[s,s2,s3,s4,s5,s6]):
            texts2.add(MathTex(t).move_to(sq))
        self.add(s,a,s2,a2,s3,a3,s4,a4,s5,a5,s6,a6,texts2)
        eq1 = MathTex("1+1=2")
        eq2 = MathTex("1+2=3")
        eq3 = MathTex("2+3=5")
        eq4 = MathTex("3+5=8")
        eqns=VGroup(eq1,eq2,eq3,eq4)
        eqns.arrange(DOWN,buff=1).next_to(VGroup(s,s2,s3,s4,s5,s6),LEFT,buff=1)
        self.camera.frame.scale(2).shift(3*UP)
        self.add(eqns)
```
