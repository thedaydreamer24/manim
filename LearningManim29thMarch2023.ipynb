```python
from manim import *
```

```python
config.media_width = "100%"
config.verbosity = "WARNING"
```

```python
%%manim -pql Learning1
class Learning1(Scene): 
    def construct(self):
        red_circle = Circle(color = RED, fill_opacity = 0.5)
        blue_square = Square(color = BLUE, fill_opacity = 0.8)
        blue_square.next_to(red_circle, LEFT)
        self.add(red_circle, blue_square)
```

```python
%%manim -pql Learning2
class Learning2(Scene): 
    def construct(self):
        ax = Axes(x_range = (-5,5), y_range = (-5,5))
        
        self.add(ax)
```

```python
%%manim -pql Learning3
class Learning3(Scene): 
    def construct(self):
        ax = Axes(x_range = (-5,5), y_range = (-5,5))
        curve = ax.plot(lambda x: (x+2)*x*(x-2)/2, color = RED)
        area = ax.get_area(curve, x_range = (-2,0))
        self.play(Create(ax, run_time = 5),Create(curve, run_time = 3))
        self.play(FadeIn(area))
        self.wait(2)
        

        
```

```python
%%manim -pql Learning4
class Learning4(Scene):
    def construct(self):
        green_square = Square(color = GREEN, fill_opacity = 0.5)
        self.play(DrawBorderThenFill(green_square))
        blue_circle = Circle(color = BLUE, fill_opacity = 0.8)
        self.play(Transform(green_square,blue_circle))
        self.wait()
        self.play(Indicate(blue_circle))
        self.play(FadeOut(blue_circle))
```

```python
%%manim -pql Learning5
class Learning5(Scene):
    def construct(self):
        plane = NumberPlane()
        self.add(plane)
        
        red_dot = Dot(color = RED)
        blue_dot = Dot(color = BLUE)
        blue_dot.next_to(red_dot, RIGHT+ DOWN )
        self.add(red_dot,blue_dot)
        
        s = Square(color = YELLOW)
        s.shift(2*UP + 4*RIGHT)
        self.add(s)
```
